```python
cr_x_train,cr_x_test,cr_y_train,cr_y_test=train_test_split(cr_x,cr_y,test_size=.2)
```

# build first model:1  logistic regression


```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(cr_x_train,cr_y_train)
```

```python
pred_y1=logmodel.predict(cr_x_test)
print(pred_y1)
tab_cr = confusion_matrix(pred_y1 , cr_y_test)
print(tab_cr)
print(accuracy_score(cr_y_test,pred_y1))
print(f1_score(cr_y_test,pred_y1))
```

# 2.decision_tree

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="gini",min_samples_split=25) # here by default tree is build using gini coefficient
dt.fit(cr_x_train,cr_y_train)
```

```python
pred_y2=dt.predict(cr_x_test)
print(pred_y2)
tab_cr = confusion_matrix(pred_y2 , cr_y_test)
print(tab_cr)
print(accuracy_score(cr_y_test,pred_y2))
print(f1_score(cr_y_test,pred_y2))
```

```python
# try for grid search
# from GridSearchCV
from sklearn.model_selection import GridSearchCV
search_dict={"criterion":["gini","entropy"],"min_samples_split":[25,30]}
grid=GridSearchCV(dt,param_grid=search_dict)
```

```python
grid.fit(cr_x_train,cr_y_train)
```

```python
grid.best_params_
```

```python
pred_y11=grid.predict(cr_x_test)
print(pred_y11)
tab_cr = confusion_matrix(pred_y11 , cr_y_test)
print(tab_cr)
print(accuracy_score(cr_y_test,pred_y11))
print(f1_score(cr_y_test,pred_y11))
```
